```python
#describe란 데이터 분석을 시작하기 전에 숫자형 데이터가 어떻게 생겼는지 빠르게 
# 요약 통계 정보를 확인하기 위해 사용한다
#갯수(count),평균(mean),표준편차(std),최솟값(min),1사분위수(25%),2사분위수(50%),3사분위수(75%),최댓값
df.describe()
```

```python
#인덱스 기준 정렬

df.sort_index()

df.sort_index(ascending=False) # ascending =  오름차순
```

```python
#값을 기준으로 정렬
#sort_values



df.sort_values(by='Height')
df.sort_values(by='Height', ascending=False)

#sort_values 는 여러개의 값을 조건으로 줄 수도 있다
#두 개의 조건을 줄 수 있다는 것은 첫번째 값이 먼저 기준이 되서 정렬을 한 뒤에 
#두 번째 값이 정렬을 한다는 뜻이다


df2 = DataFrame.iloc[:,0:11].head(10)
df2
df2.sort_values(by=['Height','Weight']) #키가 같은 사람들 안에서 weight의 차순 정렬을 할 것이다

df2.sort_values(by=['Weight','Height'],ascending=False)

# 정렬의 조건을 각각 줄 수도 있다

df2.sort_values(by=['Height','Weight'],ascending=[True,False])
```
